```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## sample 1

```python
# Generate 4 clusters of random data.
from sklearn.datasets import make_blobs

data, _ = make_blobs(n_samples=300, centers=4,
                     cluster_std=0.60, random_state=0)
```

```python
data.shape
```

```python
data
```

```python
# Plot the data
plt.scatter(data[:, 0], data[:, 1])
```

```python
# Use n_clusters=4 as the k value
# We can see from the plot above that there are 4 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
```

```python
# Fit the model to the data
kmeans.fit(data)
```

```python
import numpy as np
test_point =np.array([[0, 4]])
predicted_point = kmeans.predict(test_point)
```

```python
# Predict the clusters
predicted_clusters = kmeans.predict(data)
```

```python
#predicted_clusters
```

```python
predicted_point
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
cluster_labels=kmeans.labels_ # get cluster label of all data
print("cluster labels of points:", cluster_labels)
```

```python
#labels
```

```python
# Plot the predicted clusters to see if the model predicted the correct clusters
# This is visual validation that the model was trained correctly.
plt.scatter(data[:, 0], data[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
# get indexes of points in each cluster 
#Note: you can use these indexes in data 
index_cluster_0=np.where(cluster_labels==0)[0] # get indexes of points in cluster 0 
print("indexes of points in cluster 0:", index_cluster_0)
index_cluster_1=np.where(cluster_labels==1)[0] # get indexes of points in cluster 1
print("indexes of points in cluster 1:", index_cluster_1)
index_cluster_2=np.where(cluster_labels==2)[0] # get indexes of points in cluster 2
print("indexes of points in cluster 2:", index_cluster_2)
index_cluster_3=np.where(cluster_labels==3)[0] # get indexes of points in cluster 2
print("indexes of points in cluster 3:", index_cluster_3)
```

```python
#plot the results with clusters' label number
plt.plot(data[index_cluster_0,0],data[index_cluster_0,1],'ro') #samples in cluster 0 are red
plt.plot(data[index_cluster_1,0],data[index_cluster_1,1],'bo') #samples in cluster 1 are blue
plt.plot(data[index_cluster_2,0],data[index_cluster_2,1],'go') #samples in cluster 2 are green
plt.plot(data[index_cluster_3,0],data[index_cluster_3,1],'yo') #samples in cluster 3 are yellow
plt.title('Cluster 0: red, Cluster 1: blue, Cluster 2: green, Cluster 3: yellow')
plt.show()
```

## sample 2

```python
# Create dummy data
from sklearn.datasets import make_blobs

X, _ = make_blobs(n_samples=500, centers=6,
                            cluster_std=0.70, random_state=0)
```

```python
# Plot and show scatter
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# Create a kmeans model using k = 6
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6)

# Fit the model to the data
kmeans.fit(X)

# Use the data to predict the clusters
# save the predictions as `predicted_clusters`
predicted_clusters = kmeans.predict(X)
```

```python
# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=predicted_clusters, s=50, cmap='viridis')
plt.show()
```

```python
# Print the cluster centers and cluster labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
centers
```

```python
#labels
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)

plt.show()
```

## save and reload model

```python
import pickle
```

```python
# save the model to disk
filename = 'kmeans_model.sav'
pickle.dump(kmeans, open(filename, 'wb'))
```

```python
#reload in the above model
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.predict(X)
#print(result)
```
