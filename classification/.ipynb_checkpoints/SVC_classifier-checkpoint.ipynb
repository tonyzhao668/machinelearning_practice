```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib import style
style.use("ggplot")
```

```python
from sklearn.datasets import make_blobs
```

##  vector machine linear classifier

```python
X, y = make_blobs(n_samples=100, centers=2, random_state=0, cluster_std=.95)
plt.scatter(X[:, 0], X[:, 1], c=y, s=100, cmap="bwr");
plt.show()
```

```python
# Split data into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Fit to the training data and validate with the test data
from sklearn import svm
model = svm.SVC(kernel='linear')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
# Plot the decision boundaries
x_min = X[:, 0].min()
x_max = X[:, 0].max()
y_min = X[:, 1].min()
y_max = X[:, 1].max()

XX, YY = np.mgrid[x_min:x_max, y_min:y_max]
Z = model.decision_function(np.c_[XX.ravel(), YY.ravel()])

# Put the result into a color plot
Z = Z.reshape(XX.shape)
# plt.pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
plt.contour(XX, YY, Z, colors=['k', 'k', 'k'],
            linestyles=['--', '-', '--'], levels=[-.5, 0, .5])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', edgecolor='k', s=100)
plt.show()
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions,
                            target_names=["blue", "red"]))
```

## sample with diabetes data

```python
df = pd.read_csv("data/diabetes.csv")
df.head()
```

```python
target = df["Outcome"]
target_names = ["negative", "positive"]
```

```python
data = df.drop("Outcome", axis=1)
feature_names = data.columns
data.head()
```

```python
#from sklearn.model_selection import train_test_split
Xd_train, Xd_test, yd_train, yd_test = train_test_split(data, target, random_state=42)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
modelsvc = SVC(kernel='linear')
modelsvc.fit(Xd_train, yd_train)
```

```python
# Model Accuracy
print(f'Test Acc: {modelsvc.score(Xd_test, yd_test)}')
```

## save the trained model

```python
import pickle
```

```python
# save the model to disk
filename = 'svc_model.sav'
pickle.dump(modelsvc, open(filename, 'wb'))
```

```python
#reload in the above model
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(Xd_test, yd_test)
print(result)
```
