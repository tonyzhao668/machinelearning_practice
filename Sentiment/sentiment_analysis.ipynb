```python
# Load and prepare the dataset
import nltk
from nltk.corpus import movie_reviews
import random

documents = [(list(movie_reviews.words(fileid)), category)
              for category in movie_reviews.categories()
              for fileid in movie_reviews.fileids(category)]

random.shuffle(documents)
```

```python
# this all_words are dictionary of {word:frequency}
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())

# word_features are only the list of each unique word/punctuations from list the above dictionary
word_features = list(all_words)[:2000]
```

```python
word_features[:10]
```

```python
print(all_words['the'])
```

```python
def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words) #if in return "True" otherwise "False".
    return features
```

```python
# Train Naive Bayes classifier
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
featuresets[0]
```

```python
# Test the classifier
print(nltk.classify.accuracy(classifier, test_set))
```

```python
# Show the most important features as interpreted by Naive Bayes
classifier.show_most_informative_features(10)
```

## Split Sentence into {word : frequecy} dictionary

```python
# notice if not word_tokenized, the output will be in letter not in word 
import nltk
from nltk.probability import FreqDist

sentence='''This is is my sentence'''
tokens = nltk.tokenize.word_tokenize(sentence)
fdist=FreqDist(tokens)
```

```python
# Note fdist is a dictionary {word : frequecy}
print(fdist["is"]) 
```

```python
# notice that fdist is a dictionary, list it get the keys list.
words_list = list(fdist)
words_list
```
